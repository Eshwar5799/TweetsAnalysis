```python
import pandas as pd
import numpy as np
import re
import gensim
from nltk.tokenize.treebank import TreebankWordDetokenizer
```

```python
df = pd.read_csv("DataSet-MajorProject/FinalCleanedDB")
```

```python
data = df['Tweet']
```

```python
def depure_data(data):
    url_patttern = re.compile(r'https?://\S+|www\.\S+')
    data = url_patttern.sub(r'',data)
    
    data=re.sub('\S*@\S*\s?','',data)
    
    data=re.sub("\'","",data)
    
    return data
    
```

```python
def sent_to_words(sentences):
    for sentence in sentences:
        yield(gensim.utils.simple_preprocess(str(sentence),deacc=True))
```

```python
def detokenize(text):
    return TreebankWordDetokenizer().detokenize(text)
```

```python
temp=[]
data_to_list = data.values.tolist()
```

```python
for i in range (len(data_to_list)):
    temp.append(depure_data(data_to_list[i]))
```

```python
data_words = list(sent_to_words(temp))
```

```python
curr_data = []
for i in range(len(data_words)):
    curr_data.append(detokenize(data_words[i]))
#print(curr_data)
```

```python
cleaned_data=[]
```

```python
for i in range(len(curr_data)):
    curr=[]
    curr.append(curr_data[i])
    cleaned_data.append(curr)
```

```python
from keras.preprocessing.text import Tokenizer
from keras.preprocessing.sequence import pad_sequences
from keras import regularizers
```

```python
max_words = 35000
max_len = 800
tokenizer = Tokenizer(num_words=max_words)
tokenizer.fit_on_texts(cleaned_data)
sequences = tokenizer.texts_to_sequences(cleaned_data)
tweets = pad_sequences(sequences, maxlen=max_len)
print(tweets)
```

```python
from keras.layers import Embedding
embedding_layer = Embedding(1000, 64)
```

```python
from keras.models import Sequential
from keras import layers
from keras import regularizers
from keras import backend as K
from keras.callbacks import ModelCheckpoint
```

```python
model1 = Sequential()
model1.add(layers.Embedding(max_words, 30)) #The embedding layer
model1.add(layers.LSTM(15,dropout=0.5)) #Our LSTM layer
model1.add(layers.Dense(3,activation='softmax'))
```

```python
model1.summary()
```

```python
model1.compile(optimizer='rmsprop',loss='categorical_crossentropy', metrics=['accuracy'])
```
